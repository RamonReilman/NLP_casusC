```python
%%bash

wget -O cancer_vol1.txt https://www.gutenberg.org/cache/epub/15316/pg15316.txt
wget -O cancer_vol2.txt https://www.gutenberg.org/cache/epub/15317/pg15317.txt


cat cancer_vol1.txt cancer_vol2.txt > cancer_full.txt
```

```python
%%bash
python3 Tokenisation_jarno.py train cancer_full.txt cancer.enc --num-merges 100
```

```python
%%bash

python3 Tokenisation_jarno.py encode cancer_full.txt cancer.enc cancer.tok
```

```python
%%bash


#  3-gram
python3 Tokenisation_jarno.py ngram -n 3 -i cancer.tok -o gen_3gram.tok -l 100
```

```python
%%bash

python3 Tokenisation_jarno.py decode gen_3gram.tok cancer.enc out_3gram.txt
head -c 200 out_3gram.txt
```
