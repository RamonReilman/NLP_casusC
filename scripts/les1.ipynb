# Test script comparing tokenizer results
Taking the wikipedia page both dutch and english of cancer, copying and pasting them in a txt file and giving the model 500 merges to see what the results are. Are they comparable?

```python
from nlp_jarno import BPETokenizer
```

```python
NUM_MERGES = 500
```

```python
# Training the dutch tokenizer
nl_tokenizer = BPETokenizer()
nl_tokenizer.train("tests_texts/nl_kanker_text.txt", NUM_MERGES)
nl_tokenizer.save_encoding("nl_encoding.pkl")
```

```python
# Training the english tokenizer
en_tokenizer = BPETokenizer()
en_tokenizer.train("tests_texts/en_cancer_text.txt", NUM_MERGES)
en_tokenizer.save_encoding("en_encoding.pkl")
```
